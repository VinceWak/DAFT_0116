#### Question 1: Find the questions and answers of last month.

```python
# your code here
from stackapi import StackAPI
from datetime import datetime, timedelta


SITE = StackAPI('stackoverflow')


now = datetime.now()
last_month = now - timedelta(days=30)


from_date = int(last_month.timestamp())


questions = SITE.fetch('questions', sort='votes', fromdate=from_date)
answers = SITE.fetch('answers', sort='votes', fromdate=from_date)


print(questions)
print(answers)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
from stackapi import StackAPI
import datetime

SITE = StackAPI('stackoverflow')

now = datetime.datetime.now()
today = now.replace(hour=0, minute=0, second=0, microsecond=0)
questions = SITE.fetch('questions', min=int(today.timestamp()), tagged='python', sort='votes', minscore=5)

if questions["items"]:
    most_voted = questions['items'][0]
    print(most_voted)
else:
    print("No questions found that match the specified criteria.")
```
