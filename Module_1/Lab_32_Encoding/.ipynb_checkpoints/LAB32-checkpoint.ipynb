Create the dataframe that reflects the following table:

TransactionID | ClientID | Profession | Bank_dep | Risk | Number of credits | Revenue 

1| 231 | Self-employed | 009 | High | 2 | 30200

2| 765 | students | 005 | high | 3 |12700

3| 453 | Horeca | 007 | medium |5 |89400

4| 231 | self-employed | 009 | high | 2 | 30200

5| 892 | finance | 003 | low | 3 | 740000


```python
import pandas as pd

data = {'TransactionID': [1, 2, 3, 4, 5],
        'ClientID': [231, 765, 453, 231, 892],
        'Profession': ['Self-employed', 'students', 'Horeca', 'self-employed', 'finance'],
        'Bank_dep': [9, 5, 7, 9, 3],
        'Risk': ['High', 'high', 'medium', 'high', 'low'],
        'Number of credits': [2, 3, 5, 2, 3],
        'Revenue': [30200, 12700, 89400, 30200, 740000]}

df = pd.DataFrame(data)
df
```

```python
df['Risk']= df['Risk'].str.lower()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df_dum=pd.get_dummies(df, prefix=["col1","col2"])
df_dum.head()
```

```python
x = pd.DataFrame(df['Risk'])
x
```

```python
from sklearn.preprocessing import OrdinalEncoder

ratings_order = ['low', 'medium', 'high']

oe = OrdinalEncoder(categories = [ratings_order], dtype=int)
oe.fit(x)
risk_ord = oe.transform(x)

risk_ord
```

```python
df['risk_enc']=risk_ord
df
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['Bank_dep'])
df['Bank_dep_num'] = le.transform(df['Bank_dep']) + 1
df.head()
```

# categorical encoding : profession

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['Profession'])
df['Profession_num'] = le.transform(df['Profession']) + 1
df.headad()
```

```python
column_order = ['TransactionID', 'ClientID', 'Profession', 'Profession_num', 'Bank_dep', 'Bank_dep_num', 'Risk', 'risk_enc', 'Number of credits', 'Revenue']
df = df[column_order]
df
```
