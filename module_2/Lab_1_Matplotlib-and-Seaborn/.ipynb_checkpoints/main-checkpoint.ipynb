```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,100)
y = x*2
z = x**2

fig, ax = plt.subplots()
ax.plot(x, y, label='y = x * 2', linewidth=2, linestyle='--')
ax.plot(x, z, label='z = x ** 2', linewidth=2, linestyle='-.')
ax.legend()
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,100)
y = x*2
z = x**2

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y, label='y = x * 2', linewidth=2, linestyle='--')
ax1.legend()
ax2.plot(x, z, label='z = x ** 2', linewidth=2, linestyle='-.')
ax2.legend()
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
x = np.arange(0,100)
y = x*2
z = x**2

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax1.plot(x, y, label='y = x * 2', linewidth=2, linestyle='--')
ax1.legend()
ax1.set_title('Line Plot of y = x * 2', fontweight='bold', fontsize=12)
ax1.set_xlabel('x', fontweight='bold', fontsize=10)
ax1.set_ylabel('y', fontweight='bold', fontsize=10)

ax2.plot(x, z, label='z = x ** 2', linewidth=2, linestyle='-.')
ax2.legend()
ax2.set_title('Line Plot of z = x ** 2', fontweight='bold', fontsize=12)
ax2.set_xlabel('x', fontweight='bold', fontsize=10)
ax2.set_ylabel('z', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
x = np.linspace(0, 5, 100)
y1 = x**2
y2 = np.exp(x)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax1.plot(x, y1, label='y = x^2', linewidth=2, linestyle='--')
ax1.plot(x, y2, label='y = e^x', linewidth=2, linestyle='-.')
ax1.legend()
ax1.set_title('Line Plot with Normal Scale', fontweight='bold', fontsize=12)
ax1.set_xlabel('x', fontweight='bold', fontsize=10)
ax1.set_ylabel('y', fontweight='bold', fontsize=10)

ax2.plot(x, y1, label='y = x^2', linewidth=2, linestyle='--')
ax2.plot(x, y2, label='y = e^x', linewidth=2, linestyle='-.')
ax2.legend()
ax2.set_title('Line Plot with Logarithmic Scale', fontweight='bold', fontsize=12)
ax2.set_xlabel('x', fontweight='bold', fontsize=10)
ax2.set_ylabel('y', fontweight='bold', fontsize=10)
ax2.set_xscale('log')
ax2.set_yscale('log')

plt.tight_layout()
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_steps_by_month = fitbit.groupby('Months_encoded').median()['Steps']
fig, ax = plt.subplots(figsize=(12, 4))
median_steps_by_month.plot(kind='bar', ax=ax)
ax.set_xlabel('Month', fontweight='bold', fontsize=10)
ax.set_ylabel('Median Steps', fontweight='bold', fontsize=10)
ax.set_title('Median Steps by Month', fontweight='bold', fontsize=12)
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.savefig("figures/median_steps_by_month.png")
plt.show()

# MEDIAN STEPS BY WORK_OR_WEEKEND
median_steps_by_work_weekend = fitbit.groupby('Work_or_Weekend').median()['Steps']
fig, ax = plt.subplots(figsize=(12, 4))
median_steps_by_work_weekend.plot(kind='bar', ax=ax)
ax.set_xlabel('Work/Weekend', fontweight='bold', fontsize=10)
ax.set_ylabel('Median Steps', fontweight='bold', fontsize=10)
ax.set_title('Median Steps by Work/Weekend', fontweight='bold', fontsize=12)
plt.xticks(range(2), ['Work', 'Weekend'])
plt.savefig("figures/median_steps_by_work_weekend.png")
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit = pd.read_csv("Fitbit2.csv")

features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

fig, ax = plt.subplots(figsize=(12, 4), nrows=1, ncols=3)

for i, feature in enumerate(features):
    ax[i].scatter(fitbit[feature], fitbit['Steps'])
    ax[i].set_xlabel(feature, fontweight='bold', fontsize=10)
    ax[i].set_ylabel('Steps', fontweight='bold', fontsize=10)
    ax[i].set_title(f'{feature} vs Steps', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.savefig("figures/minutes_vs_steps.png")
plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
import matplotlib.pyplot as plt

plt.hist(titanic["Age"].dropna(), bins=30, edgecolor="k")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age")
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
fig, axis = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
axis[0].hist(titanic["Age"].dropna(), bins=10, edgecolor="k")
axis[0].set_xlabel("Age")
axis[0].set_ylabel("Count")
axis[0].set_title("Bin size: 10")

axis[1].hist(titanic["Age"].dropna(), bins=20, edgecolor="k")
axis[1].set_xlabel("Age")
axis[1].set_ylabel("Count")
axis[1].set_title("Bin size: 20")

axis[2].hist(titanic["Age"].dropna(), bins=50, edgecolor="k")
axis[2].set_xlabel("Age")
axis[2].set_ylabel("Count")
axis[2].set_title("Bin size: 50")

plt.show()
```

#### How does the bin size affect your plot?

```python
"""
The bin size affects the plot by determining the granularity of the distribution. A smaller bin size results in a plot with more bars, showing a more detailed view of the distribution, while a larger bin size results in a plot with fewer bars, giving a more general overview of the distribution
"""
```
